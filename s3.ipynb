```python
def write_parquet_file():
    df = pd.read_csv('/home/abetifihene/Desktop/blossom/companies.csv')
    df.to_parquet('/home/abetifihene/Desktop/blossom/companies.parquet')
write_parquet_file()
```

```python
company_filter.to_json('company.json', orient='records', lines=True)
company_filter.to_json('company.json.gz', orient='records', lines=True, compression='gzip')
```

```python
import boto3
region = 'eu-west-1'
s3_client = boto3.client('s3',region_name = region)
location = {'LocationConstraint':region}

bucket = 'blossom-data-eng-asiedu'
s3_client.create_bucket(Bucket=bucket,
                       CreateBucketConfiguration=location)
```

```python
s3_client.upload_file('company.json.gz', bucket, 'project2/free-7-million-company-dataset-json.gzip')
s3_client.upload_file('companies.csv', bucket,'project2/free-7-million-company-dataset.csv')
s3_client.upload_file('companies.parquet', bucket, 'project2/free-7-million-company-dataset.parquet')
```
